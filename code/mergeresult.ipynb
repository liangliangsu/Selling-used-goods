```python
path='E:\\kaggle\\Avito Demand Prediction Challenge\\features\\dataset\\'
result=pd.read_csv(path+'best_public_blend.csv',header=0)
result.columns.values[1]='r1'
data1=pd.read_csv(path+'lgsub_noCV.csv',header=0)
data1.columns.values[1]='r2'
data2=pd.read_csv(path+'resultwithwhole.001.csv',header=0)
data2.columns.values[1]='r3'

result=pd.merge(result,data1, on=['item_id'],how='left')
result=pd.merge(result,data2, on=['item_id'],how='left')

result
```

```python
result['deal_probability']=result['r1']*0.5+result['r2']*0.25+result['r3']*0.25
result
```
