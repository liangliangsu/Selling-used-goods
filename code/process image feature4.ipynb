```python
import numpy as np 
import pandas as pd 
import os
import gc
import psutil
from ast import literal_eval as make_tuple
```

```python
#path='E:\\kaggle\\Avito Demand Prediction Challenge\\image features\\image_train\\'
path='E:\\kaggle\\Avito Demand Prediction Challenge\\image features\\xcept_test\\'
filelist=os.listdir(path)
data=[]
for file in filelist:
    tempdata=pd.read_csv(path+file,header=0)
    if len(data)==0:
        data=tempdata
    else:
        data = pd.concat([data,tempdata],ignore_index=True)    
    del tempdata
    gc.collect()    
data.head()
```

```python
xcept_1,xcept_3,xcept_5=[],[],[]
xcept_type=[]
indexs=[]

for i in range(len(data)):
    temp1=data.loc[i,'xcept_1']
    temp2=data.loc[i,'xcept_2']
    temp3=data.loc[i,'xcept_3']
    temp4=data.loc[i,'xcept_4']
    temp5=data.loc[i,'xcept_5']
    
    indexs.append(data.loc[i,'indexs'])
    
    temp1=make_tuple(temp1)
    temp2=make_tuple(temp2)
    temp3=make_tuple(temp3)
    temp4=make_tuple(temp4)
    temp5=make_tuple(temp5)
    
    xcept_1.append(temp1[1])
    xcept_3.append((temp1[1]+temp2[1]+temp3[1]))
    xcept_5.append((temp1[1]+temp2[1]+temp3[1]+temp4[1]+temp5[1]))
    
    if temp1[0]=='image broken':
        xcept_type.append(0)
        print ('1. Cannot identify image file',data.loc[i,'indexs'])
    else:
        xcept_type.append(1)
        
    if (i%100000)==0:
        print ('Count is',int(i/1000),'k')
        
result = {'indexs':indexs,'xcept_1': xcept_1, 'xcept_3': xcept_3, 'xcept_5':xcept_5, 'xcept_type':xcept_type}
result = pd.DataFrame(data=result)

columns=['indexs','xcept_1','xcept_3','xcept_5','xcept_type']
result=result[columns]

result.to_csv('imagefeature_test_xcept.csv', index=False) 
result
```
