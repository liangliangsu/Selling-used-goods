```python
import numpy as np 
import pandas as pd 
import os
import gc
import psutil
from ast import literal_eval as make_tuple
```

```python
path='E:\\kaggle\\Avito Demand Prediction Challenge\\features\\test\\'
filelist=os.listdir(path)
data=[]
for file in filelist:
    tempdata=pd.read_csv(path+file,header=0)
    if len(data)==0:
        data=tempdata
    else:
        data=pd.merge(data, tempdata, on=['indexs'],how='left')  
    del tempdata
    gc.collect()    

data['type']=data['res_type']+data['xcept_type']+data['incept_type']
data=data.drop(['res_type','xcept_type','incept_type'], axis=1)
gc.collect()    
data.head()
```

```python
#label=data[['indexs','type']]

good=data[data['type']==3]
bad=data[data['type']<3]

#del data
gc.collect() 

good['incept_ratio1']=good['incept_1']/good['incept_3']
good['incept_ratio2']=good['incept_3']/good['incept_5']
good['resnet_ratio1']=good['resnet_1']/good['resnet_3']
good['resnet_ratio2']=good['resnet_3']/good['resnet_5']
good['xcept_ratio1']=good['xcept_1']/good['xcept_3']
good['xcept_ratio2']=good['xcept_3']/good['xcept_5']

good.describe()


```

```python
if len(bad)>0:
    bad = bad.reset_index(drop=True)
    meanvalue=good.describe()[1:2]
    meanvalue=pd.concat([meanvalue]*len(bad), ignore_index=True)
    meanvalue=pd.concat([bad['indexs'], meanvalue],axis=1)
    result=pd.concat([good,meanvalue],ignore_index=True)
else:
    result=good
    

columns=['indexs', 'avg_r', 'avg_g', 'avg_b', 'blurrness', 'dullness',
       'whiteness', 'avg_pixel_width', 'width', 'height', 'size', 
       'resnet_1', 'resnet_3', 'resnet_5', 'resnet_ratio1', 'resnet_ratio2',
       'incept_1','incept_3', 'incept_5', 'incept_ratio1', 'incept_ratio2',
       'xcept_1', 'xcept_3', 'xcept_5', 'xcept_ratio1', 'xcept_ratio2']
result['avg_r']=round(result['avg_r'],4)
result['avg_g']=round(result['avg_g'],4)
result['avg_b']=round(result['avg_b'],4)
result['blurrness']=round(result['blurrness'],4)
result['dullness']=round(result['dullness'],4)
result['whiteness']=round(result['whiteness'],4)
result['avg_pixel_width']=round(result['avg_pixel_width'],4)
result['width']=result['width'].astype(int)
result['height']=result['height'].astype(int)
result['size']=result['size'].astype(int)

result['resnet_1']=round(result['resnet_1'],4)
result['resnet_3']=round(result['resnet_3'],4)
result['resnet_5']=round(result['resnet_5'],4)
result['resnet_ratio1']=round(result['resnet_ratio1'],4)
result['resnet_ratio2']=round(result['resnet_ratio2'],4)

result['incept_1']=round(result['incept_1'],4)
result['incept_3']=round(result['incept_3'],4)
result['incept_5']=round(result['incept_5'],4)
result['incept_ratio1']=round(result['incept_ratio1'],4)
result['incept_ratio2']=round(result['incept_ratio2'],4)

result['xcept_1']=round(result['xcept_1'],4)
result['xcept_3']=round(result['xcept_3'],4)
result['xcept_5']=round(result['xcept_5'],4)
result['xcept_ratio1']=round(result['xcept_ratio1'],4)
result['xcept_ratio2']=round(result['xcept_ratio2'],4)

result=result[columns]
result.to_csv('imagefeature_test.csv', index=False) 
result
```
