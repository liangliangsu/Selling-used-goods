```python
import numpy as np 
import pandas as pd 
import os
import gc
import psutil
```

```python
print(os.listdir('E:\\kaggle\\Avito Demand Prediction Challenge'))
```

```python
path='E:\\kaggle\\Avito Demand Prediction Challenge\\features\\'
test=pd.read_csv(path+'imagefeature_test.csv',header=0)
train=pd.read_csv(path+'imagefeature_train.csv',header=0)

data=pd.concat([train,test],ignore_index=True)
del train,test
gc.collect()
data.head()
```

```python
data['blurrness']=np.log(data['blurrness'])
data['avg_pixel_width']=np.log(data['avg_pixel_width']+1)
data['width']=np.log(data['width'])
data['height']=np.log(data['height'])
data['size']=np.log(data['size'])

data['blurrness']=(data['blurrness']-min(data['blurrness']))/(max(data['blurrness'])-min(data['blurrness']))
data['avg_pixel_width']=(data['avg_pixel_width']-min(data['avg_pixel_width']))/(max(data['avg_pixel_width'])-min(data['avg_pixel_width']))
data['width']=(data['width']-min(data['width']))/(max(data['width'])-min(data['width']))
data['height']=(data['height']-min(data['height']))/(max(data['height'])-min(data['height']))
data['size']=(data['size']-min(data['size']))/(max(data['size'])-min(data['size']))

data.head()
```

```python
temp=pd.read_csv(path+'noimage.csv',header=0)
data=pd.concat([data,temp],ignore_index=True)
del temp
gc.collect()
data.rename(columns={"indexs": "image"}, inplace=True)
data.head()
```

```python
dataset=pd.read_csv(path+'datafeature.csv',header=0)
dataset['image'].fillna('noimage', inplace=True)
dataset['image']=(dataset['image']+'.jpg').astype(str)
dataset
```

```python
dataset['price']=dataset['price'].fillna(0)
dataset['price']=dataset['price'].clip(0,90000000)
dataset['price']=dataset['price']+1
dataset['price']=np.log(dataset['price'])
#dataset['item_seq_number']=np.log(dataset['item_seq_number'])

dataset['price']=(dataset['price']-min(dataset['price']))/(max(dataset['price'])-min(dataset['price']))
dataset.head()
```

```python
temp=dataset[dataset['deal_probability']>=0]
temp1=dataset.groupby(['item_seq_number']).item_id.size().reset_index()
temp1.columns.values[1]='item_seq_count'
temp=temp.groupby(['item_seq_number']).deal_probability.mean().reset_index()
temp.columns.values[1]='item_seq_deal_avg'
dataset=pd.merge(dataset,temp1, on=['item_seq_number'],how='left')
dataset=pd.merge(dataset,temp, on=['item_seq_number'],how='left')
dataset['item_seq_deal_avg']=dataset['item_seq_deal_avg'].fillna(-0.1)

del temp,temp1
gc.collect()
dataset
```

```python
dataset['item_seq_count']=np.log(dataset['item_seq_count'])
dataset['item_seq_count']=(dataset['item_seq_count']-min(dataset['item_seq_count']))/(max(dataset['item_seq_count'])-min(dataset['item_seq_count']))
dataset['item_seq_deal_avg']=(dataset['item_seq_deal_avg']-min(dataset['item_seq_deal_avg']))/(max(dataset['item_seq_deal_avg'])-min(dataset['item_seq_deal_avg']))
dataset
```

```python
dataset=pd.merge(dataset,data, on=['image'],how='left')
del data
gc.collect()
dataset
```

```python
columns=['item_id', 'user_id', 'title', 'description',
         'region', 'city', 'parent_category_name','category_name', 'user_type','activation_date',
         'param_1', 'param123','image_top_1','cluster_5', 'cluster_10','cluster_20','item_seq_number',         
         'para_num','item_seq_count','item_seq_deal_avg',
         'deal_prob_avg', 'deal_prob_avg_par', 'count_avg', 'count_avg_par',
         'dis_days', 'edit_time', 'item_num',
         'count',  'cus', 'ship', 'r_cus', 'r_ship', 'all_cus', 
         'avg_r','avg_g', 'avg_b', 'blurrness', 'dullness', 'whiteness',
         'avg_pixel_width', 'width', 'height', 'size', 
         'resnet_1', 'resnet_3','resnet_5', 'resnet_ratio1', 'resnet_ratio2', 
         'incept_1', 'incept_3','incept_5', 'incept_ratio1', 'incept_ratio2', 
         'xcept_1', 'xcept_3','xcept_5', 'xcept_ratio1', 'xcept_ratio2',
         'price','deal_probability']
dataset=dataset[columns]
gc.collect()
dataset
```

```python
train=dataset[dataset['deal_probability']>=0]
test=dataset[dataset['deal_probability']==-1]
test=test.drop(['deal_probability'], axis=1)
del dataset
gc.collect()
train.to_csv('trainready.csv', index=False)
test.to_csv('testready.csv', index=False)
```
