```python
import numpy as np 
import pandas as pd 
import os
import gc
import psutil
from ast import literal_eval as make_tuple
```

```python
#path='E:\\kaggle\\Avito Demand Prediction Challenge\\image features\\image_train\\'
path='E:\\kaggle\\Avito Demand Prediction Challenge\\image features\\incept_test\\'
filelist=os.listdir(path)
data=[]
for file in filelist:
    tempdata=pd.read_csv(path+file,header=0)
    if len(data)==0:
        data=tempdata
    else:
        data = pd.concat([data,tempdata],ignore_index=True)    
    del tempdata
    gc.collect()    
data.head()
```

```python
incept_1,incept_3,incept_5=[],[],[]
incept_type=[]
indexs=[]

for i in range(len(data)):
    temp1=data.loc[i,'incept_1']
    temp2=data.loc[i,'incept_2']
    temp3=data.loc[i,'incept_3']
    temp4=data.loc[i,'incept_4']
    temp5=data.loc[i,'incept_5']
    
    indexs.append(data.loc[i,'indexs'])
    
    temp1=make_tuple(temp1)
    temp2=make_tuple(temp2)
    temp3=make_tuple(temp3)
    temp4=make_tuple(temp4)
    temp5=make_tuple(temp5)
    
    incept_1.append(temp1[1])
    incept_3.append((temp1[1]+temp2[1]+temp3[1]))
    incept_5.append((temp1[1]+temp2[1]+temp3[1]+temp4[1]+temp5[1]))
    
    if temp1[0]=='image broken':
        incept_type.append(0)
        print ('1. Cannot identify image file',data.loc[i,'indexs'])
    else:
        incept_type.append(1)
        
    if (i%100000)==0:
        print ('Count is',int(i/1000),'k')
        
result = {'indexs':indexs,'incept_1': incept_1, 'incept_3': incept_3, 'incept_5':incept_5, 'incept_type':incept_type}
result = pd.DataFrame(data=result)

columns=['indexs','incept_1','incept_3','incept_5','incept_type']
result=result[columns]

result.to_csv('imagefeature_test_incept.csv', index=False) 
result
```
