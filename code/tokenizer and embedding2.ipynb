```python
import numpy as np 
import pandas as pd 
import os
import gc
import psutil
```

```python
path='E:\\kaggle\\Avito Demand Prediction Challenge\\features\\dataset\\'
test=pd.read_csv(path+'testtemp.csv',header=0)
train=pd.read_csv(path+'traintemp.csv',header=0)
test['deal_probability']=-1

data=pd.concat([train,test],ignore_index=True)
del train,test
gc.collect()
data
```

```python
from keras.preprocessing import text, sequence
all_text= (data['title']+" "+data['description']).astype(str)
all_text = np.hstack([all_text.str.lower()])
tokenizer = text.Tokenizer(num_words = 200000)
tokenizer.fit_on_texts(all_text)
del all_text
gc.collect()
```

```python
EMBEDDING_DIM1 = 300
EMBEDDING_FILE1 = 'E:\\kaggle\\Avito Demand Prediction Challenge\\ru300\\cc.ru.300.vec'
def get_coefs(word, *arr): return word, np.asarray(arr, dtype='float32')
embeddings_index1 = dict(get_coefs(*o.rstrip().rsplit(' ')) for o in open(EMBEDDING_FILE1,encoding="utf8"))
```

```python
wordlist=[]
for word, i in tokenizer.word_index.items():
    if (i<=199999):
        wordlist.append(word)
len(wordlist)        
```

```python
tokenizer = text.Tokenizer()
tokenizer.fit_on_texts(wordlist)
```

```python
data['description']=(data['title']+" "+data['description']).astype(str)
data['description']= tokenizer.texts_to_sequences(data.description.astype(str))
data['title']= tokenizer.texts_to_sequences(data.title.astype(str))
data
```

```python
vocab_size = len(tokenizer.word_index)+2
EMBEDDING_DIM1 = 300# this is from the pretrained vectors
embedding_matrix1 = np.zeros((vocab_size, EMBEDDING_DIM1))
print(embedding_matrix1.shape)
# Creating Embedding matrix 
c = 0 
c1 = 0 
w_Y = []
w_No = []
for word, i in tokenizer.word_index.items():
    if word in embeddings_index1:
        c +=1
        embedding_vector = embeddings_index1[word]
        w_Y.append(word)
    else:
        embedding_vector = None
        w_No.append(word)
        c1 +=1
    if embedding_vector is not None:    
        embedding_matrix1[i] = embedding_vector

print(c,c1, len(w_No), len(w_Y))
print(embedding_matrix1.shape)
del embeddings_index1
gc.collect()

print(" FAST TEXT DONE")


embedding_matrix1
```

```python
np.save('embedding_matrix1',embedding_matrix1)
```
