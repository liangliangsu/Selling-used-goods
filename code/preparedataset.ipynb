```python
import numpy as np 
import pandas as pd 
import os
import gc
import psutil
```

```python
print(os.listdir('E:\\kaggle\\Avito Demand Prediction Challenge'))
```

```python
#columns=['item_id','user_id','activation_date']
#columns=['item_id','user_id']
path='E:\\kaggle\\Avito Demand Prediction Challenge\\'
test=pd.read_csv(path+'test.csv',header=0)
test['deal_probability']=-1
train=pd.read_csv(path+'train.csv',header=0)
train['price'] = train['price'].fillna(0).astype('float32')
data=pd.concat([train,test],ignore_index=True)
del train,test
gc.collect()

data
```

```python
data['para_num']=(data['param_1'].isnull().astype('int')+data['param_2'].isnull().astype('int')+data['param_3'].isnull().astype('int'))
data['para_num']=(3-data['para_num'])/3.0
data['activation_date']=pd.to_datetime(data['activation_date'], format="%Y-%m-%d")
data['activation_date']=data['activation_date'].dt.weekday
data.head()
```

```python
data['param_1'].fillna(value='missing', inplace=True)
data['param_1'] = data['param_1'].astype(str)
data['param_2'].fillna(value='missing', inplace=True)
data['param_2'] = data['param_2'].astype(str)
data['param_3'].fillna(value='missing', inplace=True)
data['param_3'] = data['param_3'].astype(str)
data['user_type'].fillna(value='missing', inplace=True)
data['user_type'] = data['user_type'].astype(str)
data['image_top_1'].fillna(value='missing', inplace=True)
data['image_top_1'] = data['image_top_1'].astype(str)

data['param123'] = (data['param_1']+'_'+data['param_2']+'_'+data['param_3']).astype(str)
del data['param_2'], data['param_3']
gc.collect()
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(data['param_1'])
data['param_1'] = le.transform(data['param_1'])
print ('param_1'+' ',len(np.unique(data['param_1'])))

le = preprocessing.LabelEncoder()
le.fit(data['param123'])
data['param123'] = le.transform(data['param123'])
print ('param123'+' ',len(np.unique(data['param123'])))

le = preprocessing.LabelEncoder()
le.fit(data['user_type'])
data['user_type'] = le.transform(data['user_type'])
print ('user_type'+' ',len(np.unique(data['user_type'])))

le = preprocessing.LabelEncoder()
le.fit(data['image_top_1'])
data['image_top_1'] = le.transform(data['image_top_1'])
print ('image_top_1'+' ',len(np.unique(data['image_top_1'])))

le = preprocessing.LabelEncoder()
le.fit(data['item_seq_number'])
data['item_seq_number'] = le.transform(data['item_seq_number'])
print ('item_seq_number'+' ',len(np.unique(data['item_seq_number'])))

data.head()
```

```python
path='E:\\kaggle\\Avito Demand Prediction Challenge\\features\\'
temp=pd.read_csv(path+'classfeature.csv',header=0)
data=pd.merge(data,temp, on=['parent_category_name','category_name'],how='left')
del temp
gc.collect()
data.head()
```

```python
data['count_avg']=np.log(data['count_avg'])
data['count_avg_par']=np.log(data['count_avg_par'])

data['count_avg']=(data['count_avg']-min(data['count_avg']))/(max(data['count_avg'])-min(data['count_avg']))
data['count_avg_par']=(data['count_avg_par']-min(data['count_avg_par']))/(max(data['count_avg_par'])-min(data['count_avg_par']))

le = preprocessing.LabelEncoder()
le.fit(data['parent_category_name'])
data['parent_category_name'] = le.transform(data['parent_category_name'])
print ('parent_category_name'+' ',len(np.unique(data['parent_category_name'])))

le = preprocessing.LabelEncoder()
le.fit(data['category_name'])
data['category_name'] = le.transform(data['category_name'])
print ('category_name'+' ',len(np.unique(data['category_name'])))

data.head()
```

```python
path='E:\\kaggle\\Avito Demand Prediction Challenge\\features\\'
temp=pd.read_csv(path+'userdisplaydays.csv',header=0)
data=pd.merge(data,temp, on=['user_id'],how='left')
del temp
gc.collect()

temp=pd.read_csv(path+'usereditnum.csv',header=0)
data=pd.merge(data,temp, on=['user_id'],how='left')
del temp
gc.collect()

temp=pd.read_csv(path+'usernum.csv',header=0)
data=pd.merge(data,temp, on=['user_id'],how='left')
del temp
gc.collect()
data.head()
```

```python
data['dis_days']=data['dis_days'].fillna(1)
data['edit_time']=data['edit_time'].fillna(0)
data['item_num']=np.log(data['item_num'])

data['dis_days']=(data['dis_days']-min(data['dis_days']))/(max(data['dis_days'])-min(data['dis_days']))
data['edit_time']=(data['edit_time']-min(data['edit_time']))/(max(data['edit_time'])-min(data['edit_time']))
data['item_num']=(data['item_num']-min(data['item_num']))/(max(data['item_num'])-min(data['item_num']))
data.head()
```

```python
path='E:\\kaggle\\Avito Demand Prediction Challenge\\features\\'
temp=pd.read_csv(path+'regionfeature.csv',header=0)
data=pd.merge(data,temp, on=['region','city'],how='left')
del temp
gc.collect()
data.head()
```

```python
data['city']=(data['city']+'_'+data['region']).astype(str)
le = preprocessing.LabelEncoder()
le.fit(data['region'])
data['region'] = le.transform(data['region'])
print ('region'+' ',len(np.unique(data['region'])))

le = preprocessing.LabelEncoder()
le.fit(data['city'])
data['city'] = le.transform(data['city'])
print ('city'+' ',len(np.unique(data['city'])))

le = preprocessing.LabelEncoder()
le.fit(data['cluster_5'])
data['cluster_5'] = le.transform(data['cluster_5'])
print ('cluster_5'+' ',len(np.unique(data['cluster_5'])))

le = preprocessing.LabelEncoder()
le.fit(data['cluster_10'])
data['cluster_10'] = le.transform(data['cluster_10'])
print ('cluster_10'+' ',len(np.unique(data['cluster_10'])))

le = preprocessing.LabelEncoder()
le.fit(data['cluster_20'])
data['cluster_20'] = le.transform(data['cluster_20'])
print ('cluster_20'+' ',len(np.unique(data['cluster_20'])))

data
```
